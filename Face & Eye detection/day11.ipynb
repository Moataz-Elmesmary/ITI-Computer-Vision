```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eyeClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
```

```python
img=cv2.imread("image.jpg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12,12))
#plt.imshow(gray,cmap='gray')
faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=2,minSize=(70,70))

#print(faces)
#print(faces.shape)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
    face=gray[y:y+h,x:x+w]
    eyes=eyeClassifier.detectMultiScale(face,scaleFactor=1.01,minNeighbors=6,minSize=(70,70))
    for x2,y2,w2,h2 in eyes:
        cv2.rectangle(img,(x+x2,y+y2),(x+x2+w2,y+y2+h2),(0,255,0),5)
        #print(eyes)

#print(eyes)
plt.imshow(img[:,:,::-1])
#plt.imshow(face,cmap='gray')


```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
cap=cv2.VideoCapture(0)
faceClassifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

while (cap.isOpened()):

    ret,frame=cap.read()

    if (ret):
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        faces=faceClassifier.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=2,minSize=(70,70))
        for x,y,w,h in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),5)

    cv2.imshow("my_image",frame)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()

        
        
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

Classifier=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_fullbody.xml')

def nothing(x):
    pass

cv2.namedWindow('image')

cv2.createTrackbar('scale',"image",1,1000,nothing)
cv2.createTrackbar('neighbors',"image",1,150,nothing)
cv2.createTrackbar('size',"image",1,150,nothing)

while True:
    img=cv2.imread("pedestrian01.jpg")
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    scale=1+cv2.getTrackbarPos('scale',"image")/1000
    n=cv2.getTrackbarPos('neighbors',"image")
    size=cv2.getTrackbarPos('size',"image")
    print(scale)
    bodies=Classifier.detectMultiScale(gray,scaleFactor=scale,minNeighbors=n,minSize=(size,size))
    for x,y,w,h in bodies:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)

    cv2.imshow("image",img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break

cv2.destroyAllWindows()

    
```
