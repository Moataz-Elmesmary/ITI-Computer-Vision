## Import Packages

```python
#importing some useful packages
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

## Read in an Image

```python
#reading in an image
image = cv2.imread('rubik.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#printing out some stats and plotting
print('This image is:', type(image), 'with dimensions:', image.shape)
plt.imshow(image)  # if you wanted to show a single color channel image called 'gray', for example, call as plt.imshow(gray, cmap='gray')
```

## Sobel Edge detectors

```python
kernel0 = np.array(
    [
        [1, 0, -1],
        [2, 0, -2],
        [1, 0, -1]
    ]
, dtype=np.float32)


kernel1 = np.array(
    [
        [1, 2, 1],
        [0, 0, 0],
        [-1, -2, -1]
    ]
, dtype=np.float32)


kernel2 = np.ones((5,5),np.float32)/25
#dst1 = cv2.filter2D(img, -1, kernel1, borderType=cv2.BORDER_ISOLATED)

kernel3 = np.array(
    [
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9]
    ]
, dtype=np.float32)


out0 = cv2.filter2D(image, -1, kernel0, borderType=cv2.BORDER_ISOLATED)
out1 = cv2.filter2D(image, -1, kernel1, borderType=cv2.BORDER_ISOLATED)
out2 = cv2.filter2D(image, -1, kernel2, borderType=cv2.BORDER_ISOLATED)

out3 = cv2.filter2D(out2, -1, kernel0, borderType=cv2.BORDER_ISOLATED)
out4 = cv2.filter2D(out2, -1, kernel1, borderType=cv2.BORDER_ISOLATED)

f, ((ax00, ax01, ax02), (ax10, ax11, ax12), (ax20, ax21, ax22)) = plt.subplots(3, 3, figsize=(10,10))

ax00.imshow(image, cmap='gray')
ax00.set_title('Source')
ax01.imshow(kernel0, cmap='gray')
ax01.set_title('Kernel 0')
ax02.imshow(out0, cmap='gray')
ax02.set_title('Output 0')

ax10.imshow(image, cmap='gray')
ax10.set_title('Source')
ax11.imshow(kernel1, cmap='gray')
ax11.set_title('Kernel 1')
ax12.imshow(out1, cmap='gray')
ax12.set_title('Output 1')

ax20.imshow(image, cmap='gray')
ax20.set_title('Source')
ax21.imshow(kernel2, cmap='gray')
ax21.set_title('Kernel 2')
ax22.imshow(out2, cmap='gray')
ax22.set_title('Output 2')
```

```python
f, ((ax00, ax01, ax02), (ax10, ax11, ax12)) = plt.subplots(2, 3, figsize=(10,10))

ax00.imshow(image, cmap='gray')
ax00.set_title('Source')
ax01.imshow(out0, cmap='gray')
ax01.set_title('Kernel 0')
ax02.imshow(out3, cmap='gray')
ax02.set_title('Output 3')

ax10.imshow(image, cmap='gray')
ax10.set_title('Source')
ax11.imshow(out1, cmap='gray')
ax11.set_title('Kernel 1')
ax12.imshow(out4, cmap='gray')
ax12.set_title('Output 4')
```
