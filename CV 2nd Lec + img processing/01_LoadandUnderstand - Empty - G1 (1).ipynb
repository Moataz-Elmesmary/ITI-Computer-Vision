## Read Video or Live from CAM

```python
# read a videe 'Files/11_output.avi', then:
# loop on the video frame by frame till the end:
# -- Show the frame
# -- Then add a wait for user input
# -- Ask the user to save the image or quit
import cv2  # opencv
import numpy as np
#from matplotlib import pyplot as plt

cap = cv2.VideoCapture('Files/11_output.avi')

# Code here
while(cap.isOpened()):
    ret, frame = cap.read()
    if (ret == True):
        cv2.imshow('Frame', frame)
        k = cv2.waitKey(100) # 0 indicates infinete loop
        if k == ord('q'):
            break
        elif k == ord('s'):
            cv2.imwrite('output.png', frame)
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# similar to the above. However, change the wait to 100 and save the frame id
cap = cv2.VideoCapture('Files/11_output.avi')
#cap = cv2.VideoCapture(0) # For reading from camera

frame_count = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if (ret == True):
        frame_count += 1
        cv2.imshow('Frame', frame)
        k = cv2.waitKey(50) # 0 indicates infinete loop
        if k == ord('q'):
            break
        #elif k == ord('s'):
        cv2.imwrite(f'OutputFiles/output_{frame_count}.png', frame)
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2  # opencv
import numpy as np
# Display a gray image 500 x 500 filled with a single gray level using openCV
# First: Create a window
# Second: A numpy array with the desired value (0: black - 255:white)
def nothing(x):
    return

img = np.zeros((500, 500, 1), np.uint8)
initial_val = 100

cv2.namedWindow('Image')
cv2.createTrackbar('Trackbar', 'Image', initial_val, 255, nothing)

while (1):
    val = cv2.getTrackbarPos('Trackbar', 'Image')
    img [:] = val
    cv2.imshow('Image', img)
    k = cv2.waitKey(1) # 0 indicates infinete loop
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```
