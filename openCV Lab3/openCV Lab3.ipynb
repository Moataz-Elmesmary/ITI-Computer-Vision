# Q2 make trackbar to control line position ,color and thickness in image 500*500 rgb

```python
def nothing(x):
    pass

cv2.namedWindow('Image')

cv2.createTrackbar('x0', 'Image', 0, 500, nothing)
cv2.createTrackbar('y0', 'Image', 0, 500, nothing)
cv2.createTrackbar('x1', 'Image', 0, 500, nothing)
cv2.createTrackbar('y1', 'Image', 0, 500, nothing)
cv2.createTrackbar('b', 'Image', 0, 255, nothing)
cv2.createTrackbar('g', 'Image', 0, 255, nothing)
cv2.createTrackbar('r', 'Image', 0, 255, nothing)
cv2.createTrackbar('th', 'Image', 0, 10, nothing)

while 1:
    img = np.zeros((500, 500, 3), np.uint8)
    x0 = cv2.getTrackbarPos('x0', 'Image')
    y0 = cv2.getTrackbarPos('y0', 'Image')
    x1 = cv2.getTrackbarPos('x1', 'Image')
    y1 = cv2.getTrackbarPos('y1', 'Image')
    b = cv2.getTrackbarPos('b', 'Image')
    g = cv2.getTrackbarPos('g', 'Image')
    r = cv2.getTrackbarPos('r', 'Image')
    th = cv2.getTrackbarPos('th', 'Image')

    cv2.line(img, (x0, y0), (x1, y1), (b,g,r), th+1)
    cv2.imshow('Image', img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```

# Q3 make trackbar to control position,radius and color of circle in image 500*500 rgb.

```python
def nothing(x):
    pass

cv2.namedWindow('Image')

cv2.createTrackbar('x', 'Image', 0, 500, nothing)
cv2.createTrackbar('y', 'Image', 0, 500, nothing)
cv2.createTrackbar('r', 'Image', 1, 500, nothing)
cv2.createTrackbar('th', 'Image', 1, 100, nothing)

while 1:
    img = np.zeros((500, 500, 3), np.uint8)
    x = cv2.getTrackbarPos('x', 'Image')
    y = cv2.getTrackbarPos('y', 'Image')
    r = cv2.getTrackbarPos('r', 'Image')
    th = cv2.getTrackbarPos('th', 'Image')

    cv2.circle(img, (x, y), r, (0, 0, 255), th-1)
    cv2.imshow('Image', img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```
