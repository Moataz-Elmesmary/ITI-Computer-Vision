```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('blur.png')
plt.imshow(image)
```

```python
kernel = np.array(
    [
        [0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]
    ])
out = cv2.filter2D(image, -1, kernel)
plt.imshow(out)
```
