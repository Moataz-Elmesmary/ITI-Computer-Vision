```python
img = np.zeros((255,255), dtype = 'uint8')
```

Diagonal

```python
plt.imshow(img, cmap='gray');
```

```python
np.fill_diagonal(img, 255)
plt.imshow(img, cmap='gray');
```

```python
img2 = np.fliplr(img)
plt.imshow(img2, cmap='gray');
```

```python
img3 = np.flipud(img)
plt.imshow(img3, cmap='gray');
```
